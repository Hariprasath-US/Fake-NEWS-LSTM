```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
```

```python
vocabulary_size = 7000
```

```python
x = X.copy()
x
```

```python
x['title'][1]
```

```python
#x.drop('id',axis=1,inplace = True)
x
```

```python
x.reset_index(inplace = True)
x
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
#preprocessing

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(x)):
    rev = re.sub('[^a-zA-Z]',' ',x['title'][i])
    rev = rev.lower()
    rev = rev.split()
    rev = [ps.stem(word) for word in rev if not word in stopwords.words('english')]
    rev = ' '.join(rev)
    corpus.append(rev)
```

```python
corpus
```

```python
onehot_rep = [one_hot(words,vocabulary_size) for words in corpus]
onehot_rep
```

```python
corpus[1]
```

```python
onehot_rep[1]
```

```python
# padding 
sent_len = 20
padded_sent = pad_sequences(onehot_rep,padding = 'pre',maxlen = sent_len)
padded_sent
```

```python
len(padded_sent[1])
```

```python
 print(len(padded_sent),Y.shape)
```

```python
x_final = np.array(padded_sent)
y_final = np.array(Y)
```

```python
from sklearn.model_selection import train_test_split as TT
```

```python
x_train,x_test,y_train,y_test=TT(x_final,y_final,test_size=0.3,random_state=42)
```

```python
##Reshape targets to be 2D
#y_train = y_train.reshape(-1, 1)
#y_test = y_test.reshape(-1, 1)
# Train the model
#model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=64)
```

```python
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
vocab_size = 10000  # Size of the vocabulary (can be adjusted)
embedding_dim = 100  # Dimension for the embedding layer
max_length = 20  # Max length of input sequences
input_length = max_length

# Create the sequential model
model = Sequential()

# Add the embedding layer  
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim))

# Add LSTM layer
model.add(LSTM(100, return_sequences=False))

# Optionally, add a Dropout layer to avoid overfitting
model.add(Dropout(0.5))

# Add Dense layer for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Print the model summary
```

```python
#model trainning
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=4,batch_size=64)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = np.where(y_pred > 0.5, 1,0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
